```python
import graphlab
```

```python
# Load the CIFAR-10
```

```python
image_train = graphlab.SFrame("./data/image_train_data/")
image_test = graphlab.SFrame("./data/image_test_data/")
```

```python
# Train a nearest-neighbors model for retriving images using deep features
```

```python
knn_model = graphlab.nearest_neighbors.create(image_train,
                                             features=['deep_features'],
                                             label='id')
```

```python
# Use image retieval model with deep features to find similar images
```

```python
graphlab.canvas.set_target('ipynb')
cat = image_train[18:19]

cat['image'].show()
```

```python
q = knn_model.query(cat)
```

```python
def get_images_from_ids(query_result):
    return image_train.filter_by(query_result['reference_label'],'id')
```

```python
cat_neighbors = get_images_from_ids(knn_model.query(cat))
```

```python
cat_neighbors['image'].show()
```

```python
SFsum =image_train['label'].sketch_summary()
```

```python
dir(SFsum)
```

```python
SFsum.frequent_items().keys()
```

```python
SFrames = []
for c in SFsum.frequent_items().keys():
    SFrames.append(image_train[image_train['label']==c])
```

```python
models = []
for s in SFrames:
    knn_model = graphlab.nearest_neighbors.create(s,
                                             features=['deep_features'],
                                             label='id')
    models.append(knn_model)
    
```

```python
image_test = graphlab.SFrame("./data/image_test_data/")
```

```python
cat = image_test[0:1]
cat.show()
```

```python
q = models[3].query(cat)
```

```python
q
```

```python
cat_neighbors = get_images_from_ids(models[3].query(cat))
```

```python
cat_neighbors['image'].show()
```

```python
def get_images_from_ids(query_result):
    return image_train.filter_by(query_result['reference_label'],'id')
```
