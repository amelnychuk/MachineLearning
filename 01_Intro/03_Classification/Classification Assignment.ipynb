```python
import graphlab
```

```python
products = graphlab.SFrame('./data/amazon_baby.gl/')
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
products['word_count'] = graphlab.text_analytics.count_words(products['review'])
```

```python
products.head()
```

```python
def awe_count(wordcountdict):
    if 'awesome' in wordcountdict:
        return wordcountdict['awesome']
```

```python
products['awesome'] = products['word_count'].apply(awe_count)
```

```python
products.head()
```

```python
def wordcount(dctnry,word):
    if word in dctnry:
        return dctnry[word]
```

```python
for w in selected_words:
    fn = lambda x: x[w] if w in x else 0
    products[w] = products['word_count'].apply(fn)
```

```python
products.head()
```

# Answer 1

```python
hiscore,hiword = -1,""
lowscore,lowword = 9999999999999999,""
for w in selected_words:
    w_sum = products[w].sum()
    if w_sum > hiscore:
        hiscore,hiword = w_sum,w
    if w_sum < lowscore:
        lowscore,lowword = w_sum,w
print hiscore,hiword
print lowscore,lowword
```
